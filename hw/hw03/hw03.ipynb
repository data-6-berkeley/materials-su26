## Part 2: One Hot Encoding

Looking at the bar chart from the previous section, `"primary genre"` contains many categories, several of which have only a handful of songs. `"primary genre"` is stored as a string, which can make it tricky to do arithmatic with (such as summations). We can simplify our genres using [one-hot encoding](https://www.geeksforgeeks.org/machine-learning/ml-one-hot-encoding/) or OHE. One-hot encoding transforms our categorical data into binary variables by creating a new column for each category we care about, where each column contains a `1` if a song belongs to that category ("hot") and a `0` if it doesn't ("cold").

We'll create one-hot encoded columns for "pop", "indie", "r&b", "latin", and "rock". We're focusing on these five genres (rather than all of them) because these sub-genres appear often in `'primary genres'`. For instance, `'r&b'` appears in `'indie r&b'` and `'alternative r&b'`, so encoding `'r&b'` captures a lot of the variation in "primary genre" while keeping the number of new columns manageable. Any song whose primary genre doesn't fall into one of these five will simply have a 0 in every OHE column.
