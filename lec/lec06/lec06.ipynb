# Applying Functions

Let's revisit the `sat` table from last lecture

```python
sat = Table.read_table('data/sat2014-lecture.csv')
sat
```

```python
def avg_score(x, y, z):
    '''This function returns the average score across three SAT sections'''
    return (x + y + z) / 3
```

We can use functions of table columns with `apply`.

`.apply(function, col1, ..., coln)` will pass in the columns provided to `function`. If function takes _n_ arguments, then you should pass in _n_ columns.

```python
''' if you don't pass in any columns, and the function takes multiple arguments,
you will get an error'''
sat.apply(avg_score)
```

```python
sat.apply(avg_score, 'Critical Reading', 'Math', 'Writing')
```

```python
sat
```

```python
# This is similar to something you would see on an exam
sat.with_column('Avg. Score', sat.apply(avg_score, 'Critical Reading', 'Math', 'Writing'))
```

Let's use `apply` with some of the string functions we learned about yesterday.

```python
res_halls = Table().with_columns(
    'Residence Hall', 
        make_array('Unit 1', 'Unit 2', 'Unit 3', 'Foothill',
                   'Clark Kerr', 'Blackwell', 'Martinez Commons'),
    'Phone',
        make_array('510-642-3141', '510-642-3143', '510-642-5391', '510-642-9703',
                   '510-642-6290', '510-423-3740', '510-642-8517')
)
res_halls
```

**Task:** Convert phone numbers. For example, 510-642-3141 should be formatted as (510) 642-3141.

```python
def format_phone_number(phone):
    parts = np.array(phone.split('-'))
    return "(" + parts.item(0) + ") " + parts.item(1) + "-" + parts.item(2)

res_halls.with_columns(
    "Formatted",
    res_halls.apply(format_phone_number, "Phone")
)
```

## Scope

```python
def odd(n):
    print("value of n", n)
    return n % 2 == 1
```

```python
odd(15)
print("n outside function", n)
```

From within a function, can find and read values of names defined **outside** of function scope

```python
US_CODE = "+1"
def make_phone_number(local_phone):
    return US_CODE + local_phone
```

```python
make_phone_number("(510)642-3143")
```

```python
make_phone_number("(510)642-5391")
```
