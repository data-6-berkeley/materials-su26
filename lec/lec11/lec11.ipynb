# Control: Conditional statements

```python
def sign(x):
    """determines the sign (+/-) of a numeric input"""
    if x > 0:
        return 'Positive'
    
    elif x < 0:
        return 'Negative'
    
    elif x == 0:
        return 'Neither positive nor negative'
```

```python
sign(1)
```

```python
sign(-1)
```

```python
sign(0)
```

```python
sign('-1')
```
